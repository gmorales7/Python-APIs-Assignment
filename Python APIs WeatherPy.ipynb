```python
import csv
import pandas as pd
import requests
import json
import matplotlib.pyplot as plt
from weather import api_key
import openweathermapy as owm
import random

from citipy import citipy
```

```python
#Traditional way
url = "http://api.openweathermap.org/data/2.5/weather?"
data = requests.get(url + "appid=" + api_key +"&lat=19.432608&lon=-99.133209&units=metric").json()

# Output File (CSV)
output_data_file = "output_data/cities.csv"
```

```python
cities_list = [] # empty list for cities
info_list=[]  #  empty list for the information of the cities
counter=1 # counter to know which city number is processing

#settings, metric, in english
settings = {"units": "metric", "lang": "EN", "appid":api_key} 

#to continue until 500
while len(cities_list)<500: 
    lats = random.uniform(-90,90)  #range of random latitudes
    lons= random.uniform(-180,180)  #range of random longitudes
        
    # Identify nearest city for each (lat, lng) combination
    city = citipy.nearest_city(lats,lons).city_name 
        
    #If the city is not already in the cities list
    if city not in cities_list:
        try:  #Try to get the data of the city. For times when error appears
            data = owm.get_current(city,**settings) #Get the data of the city
            info_list.append(data("sys.country","name","coord.lat","coord.lon","main.temp","main.temp_max", "wind.speed",
                                    "clouds.all","dt","main.humidity"))  #Extract  this data and put it into the info_list
            cities_list.append(city)  #Put the city name in the already existing cities
            print("Processing city number " + str(counter) + " of 500. | City: " + cities_list[counter-1])
            counter += 1
        except:
            continue # If city is in the list, just continue the for loop
```

```python
info_list
```

```python
#Transform the list of tuples into pandas DataFrame
city_weather = pd.DataFrame(info_list, columns = ["Country_Code", "City_Name","Latitude","Longitude",'Temperature(C)',"Max_Temperature(C)",
                                  "Wind_Speed(km/h)","Clouds","Date","Humidity"])
city_weather.head()
```

```python
%matplotlib inline

ax = city_weather.plot(kind="scatter",x="Latitude", y="Max_Temperature(C)", figsize = (20,10), marker="s",color="darkblue",
                       linewidth=5)  

ax.set_facecolor("#eafff5")  #Change the background color
ax.tick_params(labelsize=14) #Change the size of the labels in the axes

plt.title("Temperature vs Latitude", fontsize = 30)  #Put a title and set the fontsize
plt.grid(b=True,color="black")  #Create a grid and make it black
plt.xlabel("Latitude", fontsize=25)  #Create an xlabel and change the fontsize
plt.ylabel("Max Temperature (C)", fontsize=25) #Create ylabel and change the fontsize
plt.xlim(-90,90)   #Set the limits for the x axis
plt.ylim(0,50)   #Set the limits for the y axis

plt.savefig("TempvsLatitudee.png")  #Save the figure
plt.show()  #Show the figure
```

We see that there's not a linear relationship between Latitude and Temperature, the further from the Equator, the smaller the temperature. We see the maximum temperatures betwteen 0' and 40'. As well, we see that in the left side of the 0, there are less points than in the right side.

```python
%matplotlib inline

ax2 = city_weather.plot(kind="scatter", x="Latitude", y="Humidity",figsize=(20,10), marker="o", color="blue",
                           linewidth=5)

ax2.set_facecolor("#eafff5")
ax2.tick_params(labelsize=14)

plt.title("Humidity(%) vs. Temperature", fontsize=30)
plt.grid(b=True, color="black")
plt.xlim(-90,90)
plt.ylim(0,105)
plt.xlabel("Latitude",fontsize = 25)
plt.ylabel("Humidity(%)", fontsize=25)

plt.savefig("HumidityTemperature.png")
plt.show()

  
```

Regarding Humidity and Latitude, there's not a clear relationship in the data; but we see that in latitudes near to 0, the humidity never seems to be low, while in latitudes more far from 0 we observe a variety of humidity conditions, but in all latitudes we see high humidities. Probably there are other factors in the cities, such as altitude or distance from the sea are impacting the humidity conditions.

```python
%matplotlib inline

ax3 = city_weather.plot(kind="scatter", x="Latitude", y="Clouds", figsize=(20,10), color="darkgreen", marker="o",
                           linewidth=5)

ax3.set_facecolor("#eafff5")
plt.title("Cloudiness vs. Latitude", fontsize = 30)
plt.grid(b=True, color="black")
plt.xlim(-90,90)
plt.ylim(-5,105)
plt.tick_params(labelsize = 18)
plt.xlabel('Latitude', fontsize = 25)
plt.ylabel("Cloudiness (%)", fontsize = 25)

plt.savefig("CloudinessLatitude.png")
plt.show()
```

The % of Latitude is clearly uncorrelated with % of cloudiness, for all latitudes we observe all different levels of cloudiness.

```python
%matplotlib inline

ax4 = city_weather.plot(kind="scatter", x= "Latitude", y="Wind_Speed(km/h)", figsize = (20,10), marker = "o", color="red",
                           linewidth=5)

ax4.set_facecolor("#eafff5")
ax4.tick_params(labelsize=18)
plt.title("Windspeed vs. Latitude", fontsize = 30)
plt.grid(b=True, color="black")
plt.xlabel("Latitude", fontsize=25)
plt.ylabel("Wind Speed (km/h)", fontsize=25)
plt.xlim(-90,90)
plt.ylim(0,20)

plt.savefig("WindspeedLatitude.png")

plt.show()
```

Wind speed and Latitude don't seem to have a linear relationship, probably near to 0; we observe high values in wind speed very rarely (also there're factors that we're not capturing with latitude), but there's not a clear pattern. The normal wind speed seems to be from low values to 7.5% for the majority of the data points.
